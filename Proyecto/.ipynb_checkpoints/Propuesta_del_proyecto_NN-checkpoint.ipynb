### Reconocimiento de emociones faciales
_Facial Emotion Recognition_ (FER) es la tecnología que analiza las expresiones faciales de imágenes estáticas y videos con el propósito de revelar información sobre el estado emocional de una persona.
Esta técnica de **análisis de sentimientos** pertenece a la familia de tecnologías a las que a menudo se hace referencia como "computación afectiva", un campo multidisciplinario de investigación sobre las capacidades de las computadoras para reconocer e interpretar las emociones humanas y los estados afectivos que, a menudo, se basa en tecnologías de inteligencia artificial. [Fuente: European Union - TechDispatch (2021)](https://www.google.com/url?sa=t&rct=j&q=&esrc=s&source=web&cd=&cad=rja&uact=8&ved=2ahUKEwiZ2_THv6f9AhXRjqQKHfsdBswQFnoECA8QAQ&url=https%3A%2F%2Fedps.europa.eu%2Fsystem%2Ffiles%2F2021-05%2F21-05-26_techdispatch-facial-emotion-recognition_ref_en.pdf&usg=AOvVaw05bM0dDN9gOWDzQL5pvIgz) 

De acuerdo al PhD. Paul Ekman, las seis emociones universales de los humanos son: felicidad, tristeza, enojo, sorpresa, miedo y disgusto. [Consultar: Ekman, P. (1970). Universal facial expressions of
emotion. California Mental Health Research Digest,
8, 151-158.](https://www.google.com/url?sa=t&rct=j&q=&esrc=s&source=web&cd=&ved=2ahUKEwjYkIHhwaf9AhUBXvEDHWXiBaIQFnoECA4QAQ&url=https%3A%2F%2Fwww.paulekman.com%2Fwp-content%2Fuploads%2F2013%2F07%2FUniversal-Facial-Expressions-of-Emotions1.pdf&usg=AOvVaw3ZOzxhvp2GEvsGfj4n8ITP).

### Propuesta de proyecto

El propósito de nuestro proyecto es desarrollar una herramienta basada en IA que clasifique las expresiones faciales en las siguientes categorías: Feliz, triste, enojado, sorprendido, asustado, disgustado y neutro. El procedimiento que pretendemos seguir es:

1. **Encontrar la cara**: Dada una imagen, encontrar la posición de la cara de la persona. Es decir, retornar la posición de la subimagen que enmarca la cara. Existen librerías en Python para esta tarea.

2. **Modelar la cara**:  A través de una red neuronal profunda, encontrar los _atributos claves_ dentro del rostro, como la posición de las cejas, la altura de los labios, entre otros. 

3. **Clasificación del estado de ánimo**: Usando los atributos clave, entrenar una red neuronal que determine la emoción predominante en la imagen. En caso de que la cantidad de atributos clave que identifiquemos sean muchos, podríamos considerar usar sólo algunos componentes principales como input.

### Título del proyecto
_"Detección de emociones faciales universales usando aprendizaje profundo"_

### Objetivos
Desarrollar un modelo de reconocimiento de expresiones faciales en Keras a través del entrenamiento de una red neuronal convolucional.

>* Comprender la teoría y la intuición detrás de las redes neuronales profundas y las redes neuronales convolucionales (CNN).
>* Diseñar y entrenar un modelo de aprendizaje profundo basado en redes neuronales convolucionales que permita identificar los atributos clave que determinan una expresión facial.
>* Entrenar una red neuronal convolucional que determine la emoción predominante a partir de atributos faciales. 
>* Aplicar el modelo a imágenes a través de una sencilla interfaz.


### Problemas
* _Capacidad computacional:_ La infraestructura para el entrenamiento de la red neuronal puede ser insuficiente. 
* _Sesgo:_ Es posible que nuestra herramienta no sea lo suficientemente sofisticada para comprender las diferencias culturales y raciales en la lectura de las emociones. Su desempeño dependerá de qué tan diverso en términos raciales es el conjunto de entrenamiento.
* _Complejidad de las emociones:_ Las expresiones faciales de las personas pueden denotar múltiples emociones de manera simultánea. 

### Datos disponibles
Para el entrenamiento de la herramienta de reconocimiento se pretende emplear algunos de los dataset disponibles en
https://paperswithcode.com/datasets?task=facial-expression-recognition&page=1

Entre ellos se encuentran **FER2013** (Facial Expression Recognition), el cual contiene aproximadamente **30.000** imágenes RGB faciales de diferentes expresiones con un tamaño restringido a 48×48, y las etiquetas principales se pueden dividir en 7 tipos: 0=Enojado, 1=Disgustado, 2=Miedo, 3=Feliz, 4=Triste, 5=Sorpresa, 6=Neutro. La expresión Disgustado tiene la cantidad mínima de imágenes: 600, mientras que otras etiquetas tienen casi 5000 muestras cada una.
